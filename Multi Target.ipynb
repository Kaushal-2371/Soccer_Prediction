```python
import pandas as pd
```

```python
table = pd.read_csv("matches.csv", index_col=0)
```

```python
table.head()
```

```python
table.dtypes
```

```python
#making date as numeric
table["date"] = pd.to_datetime(table["date"])
```

### Multidimentional result column

```python
table["result"] = table["result"].replace({"L": 0, "W": 1, "D": 2})
```

```python
table.head()
```

```python
table["opp_code"] = table["opponent"].astype("category").cat.codes
table["day_code"] = table["date"].dt.dayofweek
table["venue_code"] = table["venue"].astype("category").cat.codes
table["hour"] = table["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
# del table["comp"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code", "xg", "xga", "poss",]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50, min_samples_split=20, random_state=1)
```

```python
train = table[table["date"] < '2022-01-01']
test = table[table["date"] > '2022-01-01']
```

```python
rf.fit(train[predictors], train["result"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test["result"], preds)
accuracy
```
